```python
lotus_df = df[df['Make'] == 'Lotus']
lotus_df.head()
```

```python
lotus_df[['Engine HP', 'Engine Cylinders']].head()
```

```python
engine_hp_and_engine_cylinders_df = lotus_df[['Engine HP', 'Engine Cylinders']]
engine_hp_and_engine_cylinders_df.head()
```

```python
engine_hp_and_engine_cylinders_df.reset_index(drop=True).head()
```

```python
engine_hp_and_engine_cylinders_df = engine_hp_and_engine_cylinders_df.drop_duplicates()
```

```python
engine_hp_and_engine_cylinders_df.shape
```

```python
X = np.array(engine_hp_and_engine_cylinders_df)
X
```

```python
XT = X.T
XT
```

```python
XTX = XT.dot(X)
XTX
```

```python
XTX_inverse = np.linalg.inv(XTX)
XTX_inverse
```

```python
y = np.array([1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800])
y
```

```python
product_of_XTX_inverse_and_XT = XTX_inverse.dot(XT)
product_of_XTX_inverse_and_XT
```

```python
w = product_of_XTX_inverse_and_XT.dot(y)
w
```

the first value is **4.59494481**
